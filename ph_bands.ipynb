```python
from pymatgen.ext.matproj import MPRester
from pymatgen.phonon.plotter import PhononBSPlotter
import matplotlib.pyplot as plt
import numpy as np
from pymatgen.phonon.plotter import *
```

```python
with MPRester("Enpk9zc3KTfZxN4DTgTx") as m:
    phon = m.get_phonon_bandstructure_by_material_id("mp-30847")
plot = PhononBSPlotter(phon)
plot.get_plot()

bs_plot_data_phonon = plot.bs_plot_data()
frequency_phonon = bs_plot_data_phonon.get('frequency')
distance = bs_plot_data_phonon.get('distances')
for i in range(len(distance)): 
    plt.plot(distance[i],frequency_phonon[i][0],'green') 
    plt.plot(distance[i],frequency_phonon[i][1],'red')   
    plt.plot(distance[i],frequency_phonon[i][2],'orange')

print("Les trois directions à considérer sont Gamma-L, Gamma-K et Gamma-X.")
print("Les trois bandes acoustiques sont représentées en couleur sur le graphe, la première en vert, la seconde en rouge, et la troisième en orange.")
```

```python
# PREMIERE BANDE ACOUSTIQUE 

# Direction gamma-l : 

q1_GL = distance[4]
dq1_GL = q1_GL[2] - q1_GL[0]

f1_GL = frequency_phonon[4][0]
df1_GL = f1_GL[2] - f1_GL[0]

vs1_GL = ((df1_GL)*10**12/10**(10)*(2*np.pi))/(dq1_GL)

# Direction gamma-k : 

q1_GK = distance[3]
dq1_GK = q1_GK[-3] - q1_GK[-1]

f1_GK = frequency_phonon[3][0]
df1_GK = f1_GK[-3] - f1_GK[-1]

vs1_GK = abs(((df1_GK)*10**12/10**(10)*(2*np.pi))/(dq1_GK))

# Direction gamma-x : 

q1_GX = distance[0]
dq1_GX = q1_GX[2] - q1_GX[0]

f1_GX = frequency_phonon[0][0]
df1_GX = f1_GX[2] - f1_GX[0]

vs1_GX = ((df1_GX)*10**12/10**(10)*(2*np.pi))/(dq1_GX)

print("La vitesse du son pour la première bande acoustique dans les différentes directions est : ")
print(" - Gamma-L : ", vs1_GL)
print(" - Gamma-K : ", vs1_GK)
print(" - Gamma-X : ", vs1_GX)
```

```python
# DEUXIEME BANDE ACOUSTIQUE 

# Direction gamma-l : 

q2_GL = distance[4]
dq2_GL = q2_GL[2] - q2_GL[0]

f2_GL = frequency_phonon[4][1]
df2_GL = f2_GL[2] - f2_GL[0]

vs2_GL = ((df2_GL)*10**12/10**(10)*(2*np.pi))/(dq2_GL)

# Direction gamma-k : 

q2_GK = distance[3]
dq2_GK = q2_GK[-3] - q2_GK[-1]

f2_GK = frequency_phonon[3][1]
df2_GK = f2_GK[-3] - f2_GK[-1]

vs2_GK = abs(((df2_GK)*10**12/10**(10)*(2*np.pi))/(dq2_GK))

# Direction gamma-x : 

q2_GX = distance[0]
dq2_GX = q2_GX[2] - q2_GX[0]

f2_GX = frequency_phonon[0][1]
df2_GX = f2_GX[2] - f2_GX[0]

vs2_GX = ((df2_GX)*10**12/10**(10)*(2*np.pi))/(dq2_GX)

print("La vitesse du son pour la deuxième bande acoustique dans les différentes directions est : ")
print(" - Gamma-L : ", vs2_GL)
print(" - Gamma-K : ", vs2_GK)
print(" - Gamma-X : ", vs2_GX)
```

```python
# TROISIEME BANDE ACOUSTIQUE 

# Direction gamma-l : 

q3_GL = distance[4]
dq3_GL = q3_GL[2] - q3_GL[0]

f3_GL = frequency_phonon[4][2]
df3_GL = f3_GL[2] - f3_GL[0]

vs3_GL = ((df3_GL)*10**12/10**(10)*(2*np.pi))/(dq3_GL)

# Direction gamma-k : 

q3_GK = distance[3]
dq3_GK = q3_GK[-3] - q3_GK[-1]

f3_GK = frequency_phonon[3][2]
df3_GK = f3_GK[-3] - f3_GK[-1]

vs3_GK = abs(((df3_GK)*10**12/10**(10)*(2*np.pi))/(dq3_GK))

# Direction gamma-x : 

q3_GX = distance[0]
dq3_GX = q3_GX[2] - q3_GX[0]

f3_GX = frequency_phonon[0][2]
df3_GX = f3_GX[2] - f3_GX[0]

vs3_GX = ((df3_GX)*10**12/10**(10)*(2*np.pi))/(dq3_GX)

print("La vitesse du son pour la troisième bande acoustique dans les différentes directions est : ")
print(" - Gamma-L : ", vs3_GL)
print(" - Gamma-K : ", vs3_GK)
print(" - Gamma-X : ", vs3_GX)
```
