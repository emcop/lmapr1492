```python
from pymatgen.ext.matproj import MPRester
from pymatgen import Structure
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
from pymatgen.analysis.diffraction.core import AbstractDiffractionPatternCalculator, DiffractionPattern
from pymatgen.analysis.diffraction.xrd import XRDCalculator
from pymatgen.analysis.diffraction.tem import TEMCalculator
```

```python
with MPRester("Enpk9zc3KTfZxN4DTgTx") as m:
    structure = m.get_structure_by_material_id("mp-30847")
```

```python
XRD_calculator = XRDCalculator(wavelength='CuKa', symprec=0, debye_waller_factors=None)
pattern = XRD_calculator.get_pattern(structure = structure, scaled=True, two_theta_range=(0, 90))
XRD_calculator.show_plot(structure = structure, annotate_peaks = True)
```

```python
# PIC 1
print("Premier pic : ")
print("Intensité relative : ", pattern.y[0])
print("Angle associé : ", pattern.x[0]/2)
print("Indices hlk associés : (1, 0, 0), (1, 1, 1)")
```

```python
# PIC 2
print("Deuxième pic : ")
print("Intensité relative : ", pattern.y[1])
print("Angle associé : ", pattern.x[1]/2)
print("Indices hlk associés : (1, 1, 0)")
```

```python
# PIC 3
print("Troisième pic : ")
print("Intensité relative : ", pattern.y[2])
print("Angle associé : ", pattern.x[2]/2)
print("Indices hlk associés : (1, 0, -1), (2, 1, 1)")
```
